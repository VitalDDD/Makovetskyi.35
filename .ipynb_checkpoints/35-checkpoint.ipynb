```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import sklearn
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Перше завдання. Модель простої регресії для знаходження залежності між значеннями 
# протяжності навчання і середнього балу студентів

# Читаємо файл з даними студетів
df_student=pd.read_csv('student_scores.csv')

# Готуємо дані для аналізу 
X = df_student['Hours']
y = df_student['Scores']
X_array = np.array(X).reshape(-1, 1)
y_array = np.array(y)

# Створюємо модель з даних та отримуємо коефіцієнти прямої 
LinReg = LinearRegression()
LinReg.fit(X_array, y_array)
b, m = LinReg.intercept_, LinReg.coef_ 
m = m[0]

# Візуалізуємо дані та лінійну регресію
fig, ax = plt.subplots()
plt.scatter(df_student['Hours'], df_student['Scores'], color = 'green')
plt.plot(list(range(1, 12)), [m * x + b for x in range(1, 12)], color = 'blue')
ax.set_xlabel("Hours")
ax.set_ylabel("Scores")
ax.set_title("Залежність між протяжністю навчання і середнім балом студентів")

# Оцінюємо якість отриманої регресії
y_predict = LinReg.predict(X_array)

mae = mean_absolute_error(y_array, y_predict)
mse = mean_squared_error(y_array, y_predict)
rmse = np.sqrt(mse)

print(f'Mean Absolute Error: {mae:.2f}')
print(f'Mean Squared Error: {mse:.2f}')
print(f'Root Mean Squared Error: {rmse:.2f}')
```

```python
# Друге завдання. Mодель множинної регресії для знаходження залежності між значеннями 
# споживання бензину і похідними факторами

# Читаємо файл з даними студетів
df_petrol=pd.read_csv('petrol_consumption.csv')

# Готуємо дані для аналізу 
X_array = np.array(df_petrol.iloc[:, :-1])
y_array = np.array(df_petrol.iloc[:, -1])

# Створюємо модель з даних
LinReg = LinearRegression()
LinReg.fit(X_array, y_array)

# Оцінюємо якість отриманої регресії
y_predict = LinReg.predict(X_array)

mae = mean_absolute_error(y_array, y_predict)
mse = mean_squared_error(y_array, y_predict)
rmse = np.sqrt(mse)
print(f'Mean Absolute Error: {mae:.2f}')
print(f'Mean Squared Error: {mse:.2f}')
print(f'Root Mean Squared Error: {rmse:.2f}')

# Візуалізуємо дані та регресію
plt.scatter(y_array, y_predict)
plt.plot([y_array.min(), y_array.max()], [y_array.min(), y_array.max()], 'k--', lw=2)
plt.xlabel('Дійсні значення')
plt.ylabel('Прогнозовані значення')
plt.title('Графік прогнозованих та дійсних значень')
plt.show()
```
